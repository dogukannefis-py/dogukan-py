```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline

dataset = pd.read_csv('C:/Users/dnefis/Downloads/Weather.csv')
dataset.shape
dataset.describe()                          #istatiksel olarak gösterim


X = dataset['MinTemp'].values.reshape(-1,1)    #degerleri matrixe oturtuoz,numpy
y = dataset['MaxTemp'].values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

regressor=LinearRegression()
regressor.fit(X_train, y_train)              #algoritmayı train degerleri ögretiyoruz, bir denklem cıkarıyo x ve y ye göre

#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)

y_pred=regressor.predict(X_test)   #algoritmayı egittik,predict yapmasını istiyebiliriz, x e göre y yi tahmin ettiriyoruz

df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df1=df.head(25)
df1
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()


plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()

```
